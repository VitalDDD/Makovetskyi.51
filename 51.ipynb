```python
import pandas as pd
import numpy as np
from sklearn. model_selection import train_test_split
from sklearn import metrics
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Формуємо дані

N1 = 8000
N2 = 5000
N = N1+N2
mu_1 = 13.0
sigma_1 = 0.5
mu_2 = 12.0
sigma_2 = 0.7

X = np.zeros((N))
np.random.seed(seed=42)
X[:N1] = np.random.normal(loc=mu_1, scale=sigma_1, size=N1)
X[N1:N] = np.random.normal(loc=mu_2, scale=sigma_2, size=N2)

y = np.zeros((N))
y[:N1] = np.array((1))
y[N1:N] = np.array((0))

X_train, X_test, y_train, y_test = train_test_split (X, y, train_size=0.8, random_state=0)
X_train = X_train.reshape(-1, 1)
y_train = y_train.reshape(-1, 1)
X_test = X_test.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)
```

```python
# Модель
gm = GaussianMixture(n_components=2).fit(X_train, y_train)
y_pred = gm.predict(X_test)

cnf_matrix = metrics.confusion_matrix (y_test, y_pred)
ax = plt.subplot()
sns.heatmap(cnf_matrix, annot=True, fmt='g', ax=ax)

# Метрики
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

print("\nPrecision:",metrics.precision_score(y_test, y_pred))

print("\nRecall:",metrics.recall_score(y_test, y_pred))

print("\nF1:",metrics.f1_score(y_test, y_pred, average = None))
```
